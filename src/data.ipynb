```python
import pandas as pd
import os
dir_name="C:/Users/namra/Practicum Dataset/user_data/user01/1598759880/mMag"
dir = os.listdir( dir_name )
df_list = []

for file in dir:
   df = pd.read_csv(dir_name+"/"+file)
   df['timestamp_s'] = float(os.path.splitext(file)[0])
   df = df.rename(columns={'x': 'mMag_x','y': 'mMag_y','z': 'mMag_z'})
   df.to_csv(dir_name+"/"+file,index=False)
   df_list.append(df)

df_csv_append = pd.concat(df_list, ignore_index=True)
df_csv_append.to_csv("C:/Users/namra/Practicum Dataset/user_data/user01/1598759880/1598759880_mMag.csv",index=False)
```
